```python
#import dependencies
import pandas as pd
```

```python
#open clean data file
file="../Resources/b_corp_clean.csv"
clean_df=pd.read_csv(file, index_col="company_id")
```

```python
clean_df.head()
```

```python
#Create df by size
small_business_df =clean_df[clean_df["size"]=="1-9"]
small_business_df
```

```python
#Create df by size
med_business_df=clean_df[clean_df["size"]=="10-49"]
med_business_df
```

```python
#Create df by size
large_business_df =clean_df[clean_df["size"]=="50-249"]
large_business_df
```

```python
#Create df by size
xlarge_business_df =clean_df[clean_df["size"]=="250-999"]
xlarge_business_df
```

```python
#Create df by size
xxlarge_business_df =clean_df[clean_df["size"]=="1000+"]
xxlarge_business_df
```

```python
#sort by times recertified
xxlarge_business_df.groupby(["company_name", "date_first_certified"])["company_name"]\
.count().reset_index(name = "count").sort_values(["count"], ascending=False)
```

```python
xxlarge_business_df["company_name"].tolist()
```

```python
cascade_df = xxlarge_business_df.loc[xxlarge_business_df["company_name"]=="Cascade Engineering"]
cascade_df
```

```python
#sort by country
xxlarge_business_df.groupby(["country"])["country"]\
.count().reset_index(name = "count").sort_values(["count"], ascending=False)
```

```python
#Create df by size
unknown_business_df =clean_df[clean_df["size"]=="Unknown"]
unknown_business_df
```

```python
#calculate some percentages
small_percent = (small_business_df["current_status"]=="Certified").count()/(clean_df["current_status"]=="Certified").count()*100
small_percent=small_percent.round(decimals=2)
print(f"{small_percent}% of all Certified B Corporations are small businesses.")
med_percent = (med_business_df["current_status"]=="Certified").count()/(clean_df["current_status"]=="Certified").count()*100
med_percent=med_percent.round(decimals=2)
print(f"{med_percent}% of all Certified B Corporations are medium businesses.")
unknown_percent = (unknown_business_df["current_status"]=="Certified").count()/(clean_df["current_status"]=="Certified").count()*100
unknown_percent=unknown_percent.round(decimals=2)
print(f"{unknown_percent}% of all Certified B Corporations are of unknown size.")
```
