```python
#Import the dependencies
import pandas as pd
import re
```

```python
#import the file
file="../Resources/B Corp Impact Data.csv"
df=pd.read_csv(file, index_col="company_id")
```

```python
#examine the data types
df.dtypes
```

```python
#examine the columns
df.columns
```

```python
#fill nans with 0
df=df.fillna("0", axis=0)
df
```

```python
#create df with relevant columns
df=pd.DataFrame(df, columns=['company_name', 'date_first_certified',
       'current_status', 'industry', 'industry_category',
       'products_and_services', 'country', 'state','sector', 'size',
        'assessment_year','overall_score', 'impact_area_community','impact_area_customers', 
        'impact_area_environment', 'impact_area_governance', 'impact_area_workers','certification_cycle'])
df
```

```python
#examine dtypes of remaining columns
pd.set_option('display.max_rows', 100)
df.dtypes
```

```python
#converting date columns from object to date
df["date_first_certified"]=df["date_first_certified"].astype("datetime64")
df["date_first_certified"].dtype
```

```python
#search for double company id entries
df.index.nunique()
```

```python
#evaluate quality of column data
print(df["industry"].unique())
df["industry"].nunique()
```

```python
#evaluate quality of column data

df["company_name"].value_counts()
```

```python
#evaluate quality of column data
df["industry_category"].value_counts()
```

```python
# combine industry categories that are similar under one heading
df["industry_category"]=df["industry_category"].str.replace("Legal","Legal Services")

df["industry_category"]=df["industry_category"].str.replace("Legal Services Services","Legal Services")
df["industry_category"].value_counts()
```

```python
# combine industry categories that are similar under one heading
df["industry_category"]=df["industry_category"].str.replace("Health","Health & Human Services")

df["industry_category"]=df["industry_category"].str.replace("Health & Human Services & Human Services","Health & Human Services")
df["industry_category"].value_counts()
```

```python
#clean data to remove "*" value
df["industry_category"]=df["industry_category"].replace("*", "0")
df["industry_category"].value_counts()
```

```python
#remove 0 string values from industry category fillna earlier
df=df.loc[(df["industry_category"]!="0")].copy()
df
```

```python
#how many of the companies are currently certified vs uncertified
df["current_status"].value_counts()
```

```python
#convert all text columns to title case
# opportunity for automization (function)

df["company_name"]=df["company_name"].str.title()
df["state"]=df["state"].str.title()
df["current_status"]=df["current_status"].str.title()
df["industry"]=df["industry"].str.title()
df["industry_category"]=df["industry_category"].str.title()
df["products_and_services"]=df["products_and_services"].str.title()
df["country"]=df["country"].str.title()
df["sector"]=df["sector"].str.title()

df
```

```python
#investigate unique countrys and check for written errors and doubles
print(df["country"].nunique())
df["country"].unique()
```

```python
#check for index misspellings and correct "vietnam" duplicate
df["country"]=df["country"].str.replace("Viet Nam", "Vietnam")
df["country"].value_counts().sort_index()
```

```python
#Where are most bcorps located?
df["country"].value_counts()
```

```python
#investigate state column
#quite a lot of duplicates and unknown characters/same names in native language/english
#create regex funct to group?
df["state"].value_counts().sort_index()
```

```python
df["state"]=df["state"].str.replace("(^\d)", "Unknown")
df["state"]=df["state"].str.replace("(\?\?)", "Unknown")
df["state"].value_counts().sort_index()
```

```python
#see when first and last certifications were entered
print(min(df["date_first_certified"]))
max(df["date_first_certified"])
```

```python
#investigate size and create bins
df["size"].value_counts()
```

```python
#binning size
df["size"]=df["size"].str.replace("250[+]", "250-999")
df["size"].value_counts()
```

```python
#binning outlier size
df["size"]=df["size"].str.replace("10,000[+]", "1000+")
df["size"].value_counts()
```

```python
#who are the size 0/null/unknown businesses, should they be dropped?
size_0=df.loc[df["size"]=="0"]
print(size_0["current_status"].value_counts())
size_0.nunique()
```

```python
#keep size 0 business, rename them as "unknown" size
df["size"]=df["size"].str.replace("(^0)", "Unknown")
df["size"].value_counts()
```

```python
#inspect sector values
df["sector"].value_counts()
```

```python
#inspect products and services for different values
print(df["products_and_services"].value_counts())
df["products_and_services"].nunique()
```

```python
#this column would either need extensive binning or can be left in as description for visualization
pd.set_option('display.max_rows', 10000)
df["products_and_services"]=df["products_and_services"].str.replace("(^0)", "Unknown")
df["products_and_services"].value_counts().sort_index()
```

```python
print(df["overall_score"].min())
df["overall_score"].max()
```
